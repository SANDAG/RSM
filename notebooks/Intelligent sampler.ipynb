```python
def intel_sample(
    input_dir,
    output_dir,
    prev_iter_access = prev_access_df, 
    curr_iter_access = current_access_df,
    study_area = project_gdf,
    input_household = hh_df,
    input_person = person_df,
    compare_access_columns = access_columns,
    default_sampling_rate,
    lower_bound_sampling_rate = lb_sampling_rate, #fix the values of this after some testing
    upper_bound_sampling_rate = ub_sampling_rate #fix the values of this after some testing
):
    
    if os.path.isfile(os.path.join(input_dir, prev_iter_access)):
        prev_iter_access = pd.read_csv(os.path.join(input_dir, "accessibilities_mockup.csv"))
    else:
        raise Exception(f"The file {prev_iter_access} does not exist")
        
    #changing the taz and mgra to new cluster ids
    input_household_df = pd.read_csv(os.path.join(input_dir, input_household))
    input_household_df["taz"] = input_household_df["taz"].map(dict_clusters)
    input_household_df["mgra"] = input_household_df["taz"]
    input_household_df["count"] = 1
    
    taz_hh = input_household_df.groupby(["taz"])["count"].count().reset_index()
    
    if not os.path.isfile(os.path.join(curr_iter_access)):
    
        # true for first iteraion
        # sample 100% of households from study region
        # sample 25% of the households from rest of the region 

        taz_hh["sampling_rate"] = deafult_sampling_rate
        taz_hh.loc[taz_hh["taz"].isin(study_area), "sample_rate"] = 1

        sample_households = []

        for row in taz_hh.rows():
            df = input_household_df.loc[input_household["taz"] == row["taz"]]
            sampling_rate = row["sampling_rate"]
            df = df.sample(frac = default_sampling_rate, random_state = 1234)
            sample_households.append(df)

        #combine study are and non-study area households into single dataframe
        sample_households_df = pd.concat(sample_households)
        sample_households_df.to_csv(os.path.join(output_dir, "sample_households.csv"), index = False)
    
    else:
        #true for second, third and fourth iteration
        curr_iter_access = pd.read_csv(os.path.join(input_dir, curr_iter_access))

        #compare accessibility columns
        compare_results = pd.DataFrame()

        for column in compare_access_columns.append("MGRA"):
            if column == "MGRA":
                compare_results[column] = prev_iter_access[column]
            else:
                compare_results[column] = (curr_iter_access[column] - prev_iter_access[column]).abs() # take absolute difference

        #Take row sums of all difference    
        compare_results["Total"] = compare_results[cols_to_sum].sum(axis = 1) #

        #
        compare_results['sampling_rate'] = compare_results["Total"].clip(lower_bound_sampling_rate, upper_bound_sampling_rate)

        sample_households = []
        sample_rate_df = compare_results[["MGRA", "sampling_rate"]]
        sample_rate_df = sample_rate_df["sampling_rate"]/100
        sample_rate_df.loc[sample_rate_df["MGRA"].isin(study_area), "sampling_rate"] = 1

        for row in sample_rate_df.rows():
            df = input_household_df.loc[input_household["taz"] == row["MGRA"]]
            sampling_rate = row["sampling_rate"]
            df = df.sample(frac = default_sampling_rate, random_state = 1234)
            sample_households.append(df)

        #combine study are and non-study area households into single dataframe
        sample_households_df = pd.concat(sample_households)
        sample_households_df.to_csv(os.path.join(output_dir, "sample_households.csv"), index = False)
        
        
    #select persons belonging to sampled households
    sample_hhids = list(sample_households_df["hhid"])
    persons_df = pd.read_csv(os.path.join(input_dir, input_person))
    sample_persons_df = persons_df.loc[persons_df["hhid"].isin(sample_hhids)]
    sample_persons_df.to_csv(os.path.join(output_dir, "sample_persons.csv"), index = False)
```
